```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
data_path = os.path.join(os.getcwd(), 'OIDv4_ToolKit', 'OID', 'csv_folder')
data_path
```

```python
os.listdir(data_path)
```

```python
# only keep images that are relevant
class_desc = pd.read_csv(os.path.join(data_path,os.listdir(data_path)[0]), header = None)
class_desc.columns = ['Code', 'Label']

test = pd.read_csv(os.path.join(data_path,os.listdir(data_path)[1]))
train = pd.read_csv(os.path.join(data_path,os.listdir(data_path)[2]))
val = pd.read_csv(os.path.join(data_path,os.listdir(data_path)[3]))
```

```python
class_desc
```

```python
airbnb_classes = ['Bathtub', 'Bed', 'Billiard table', 'Ceiling fan', 'Coffeemaker', 'Couch', 'Countertop', 'Dishwasher',
                  'Fireplace', 'Fountain', 'Gas stove', 'Jacuzzi', 'Kitchen & dining room table', 'Microwave oven',
                  'Mirror', 'Oven', 'Pillow', 'Porch', 'Refrigerator', 'Shower', 'Sink', 'Sofa bed', 'Stairs', 
                  'Swimming pool', 'Television', 'Toilet', 'Towel', 'Tree house', 'Washing machine', 'Wine rack']

airbnb_classes = class_desc[class_desc['Label'].isin(airbnb_classes)]
airbnb_classes
```

```python
train = train[train['LabelName'].isin(airbnb_classes['Code'])]
test = test[test['LabelName'].isin(airbnb_classes['Code'])]
val = val[val['LabelName'].isin(airbnb_classes['Code'])]
```

```python
train.head()
```

```python
train = pd.merge(train, class_desc, left_on = 'LabelName', right_on = 'Code')
test = pd.merge(test, class_desc, left_on = 'LabelName', right_on = 'Code')
val = pd.merge(val, class_desc, left_on = 'LabelName', right_on = 'Code')

train.drop(['Code'], axis = 1, inplace = True)
test.drop(['Code'], axis = 1, inplace = True)
val.drop(['Code'], axis = 1, inplace = True)
```

```python
train.head()
```

```python
train.to_csv('train_labels.csv', encoding='utf-8', index=False)
test.to_csv('test_labels.csv', encoding='utf-8', index=False)
val.to_csv('val_labels.csv', encoding='utf-8', index=False)
```
